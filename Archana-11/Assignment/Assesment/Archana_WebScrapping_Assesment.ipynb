```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
QUOTES = []
AUTHOR = []
TAGS = []
ABOUT_AUTHOR = []
for i in range(1,11):
  re = requests.get("http://quotes.toscrape.com/page/"+str(i))
  data = BeautifulSoup(re.text , "html.parser")
  # print(data.prettify())
  quotes = data.find_all("span" , class_="text")
  for quote in quotes:
   QUOTES.append(quote.text)
  author = data.find_all("small" , class_="author")
  for auth in author:
    AUTHOR.append(auth.text)
  tags = data.find_all("meta" , class_="keywords")
  for tag in tags:
    TAGS.append(tag.attrs["content"])
  about = data.find_all("a")
  for ab in about:
    if "/author/" in ab.attrs["href"]:
      # print(ab.attrs["href"])
      ABOUT_AUTHOR.append(ab.attrs["href"])

len(QUOTES)
```
